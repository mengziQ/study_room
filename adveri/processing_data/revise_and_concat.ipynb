```python
# 欠損データ数調査
file_list = ['corpus_yasuda_2.tsv', 'corpus_prism3_20180509.tsv']
for f in file_list:
    with open('./data/{}'.format(f), 'r') as tsv:
        obj = tsv.readlines()
        cnt = 0
        for i, o in enumerate(obj):
            splited = o.split('\t')
            if len(splited) != 2:
                cnt += 1
                print(i)
                print(splited[0])
                print(splited[1])
                print(splited[2])
        print('{}: '.format(f), cnt)
```

```python
# 欠損データ修正(コーパス内にタブが入っているデータについて、タブを除去)
#file_list = ['corpus_yasuda', 'corpus_prism3_20180509']
file_list = ['corpus_pointsite']
for f in file_list:
    with open('./data/{}.tsv'.format(f), 'r') as tsv:
        with open('./data/{}_2.tsv'.format(f), 'w') as tsv2:
            obj = tsv.readlines()
            for c, o in enumerate(obj):
                splited = o.split('\t')
                if len(splited) == 2:
                    corpus = splited[0] + '\t' + splited[1].replace(' ', '')
                    tsv2.write(corpus)
                    #print(o)
                else:
                    if len(splited) < 2:
                        continue
                    corpus = ''
                    for i in range(1, len(splited)):
                        if splited[i] == '':
                            continue
                        corpus += splited[i].replace(' ', '').replace('\t', '')
                    corpus = splited[0] + '\t' + corpus
                    tsv2.write(corpus)
                    #print(corpus)
```
